```python
# load the dataset
file = cwd+'\\cust_merch.xlsx'
df_cust_merch = pd.read_excel(file)
df_cust_merch.head(10)
```

```python
# Convert the running list of user ratings into a matrix using the 'pivot table' function
ratings_df_cust_merch = pd.pivot_table(df_cust_merch, index='customer_id', columns='merchant_name', aggfunc=np.sum)
```

```python
ratings_df_cust_merch.head()
```

```python
ratings_df_cust_merch1 = ratings_df_cust_merch.iloc[:500, :500]
```

```python
ratings_df_cust_merch1.shape
```

```python
ratings_df_cust_merch1.head()
```

```python
import matrix_factorization_utilities
# Apply matrix factorization to find the latent features
U, M = matrix_factorization_utilities.low_rank_matrix_factorization(ratings_df_cust_merch1.as_matrix(),
                                                                    num_features=15,
                                                                    regularization_amount=0.1)
```

```python
print(M)
```

```python
M.shape
```

```python
# Find all predicted ratings by multiplying the U by M
predicted_ratings = np.matmul(U, M)
```

```python
import pickle
# Save features and predicted ratings to files for later use
pickle.dump(U, open("user_features.dat", "wb"))
pickle.dump(M, open("product_features.dat", "wb"))
pickle.dump(predicted_ratings, open("predicted_ratings.dat", "wb" ))
```

```python
# Save all the ratings to a csv file
predicted_ratings_df = pd.DataFrame(index=ratings_df_cust_merch1.index,
                                    columns=ratings_df_cust_merch1.columns,
                                    data=predicted_ratings)
predicted_ratings_df.to_csv("predicted_ratings.csv")
```
