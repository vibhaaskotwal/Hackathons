```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_Cust = pd.read_csv('.\\customer\\TEST1.csv')
#df_test = pd.read_csv('.\tran\TEST1.csv')
df_Tran = pd.read_csv('.\\customer\\transaction.csv')
df_Cust.head()
```

```python
# Marital Status , Gender, Card Type
gender_num = {'M': 0,'F': 1 }
df_Cust['gender'] = df_Cust['gender'].map(gender_num)
Mar_Status ={'M':1,'S':0,'D':2}
df_Cust['mar_status'] = df_Cust['mar_status'].map(Mar_Status)

df_Cust.head()
```

```python
Card_Type ={'PLATINUM':3,'GOLD':2,'SILVER':1,'WORLD':0}
df_Cust['card_type'] = df_Cust['card_type'].map(Card_Type)
```

```python
df_Tran.head()
```

```python
import datetime
# weekdays as a tuple

weekDays = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")

df_Tran['tran_date'] = pd.to_datetime(df_Tran['tran_date'])
```

```python
#pd.DatetimeIndex(df_Tran['tran_date']).day
print(df_Tran.head(5)['tran_date'].dt.weekday_name)

df_Tran["WeekDay"] = df_Tran['tran_date'].apply(lambda x: x.weekday())
# print(pd.to_datetime(df_Tran['tran_date']).weekday_name)
df_Tran.head()
```

```python
df_complete = pd.merge(df_Tran, df_Cust, on='customer_id', how='inner')
df_complete.head()
```

```python
print(df_complete.columns)
```

```python
# Build columns list and loop through it. Age, Gender,City, Card Type, MCC code type, marital_status
Columns =['age','gender','customer_city','customer_country','merchant_country','card_type','mcc_code','mar_status']
```

```python
for i in Columns :
   # df_spent_by_col = df_complete.groupby(i)["tran_amount"].sum().to_frame().reset_index()
    
    fig, ax = plt.subplots(figsize=(20,5))

    plt.title(str(i)+"/Sum of Trans_amount")
    plt.xlabel(i)
    plt.ylabel("Trans Amount")
    ax = sns.barplot(x=i, y='tran_amount',data= df_complete.groupby(i)["tran_amount"].sum().to_frame().reset_index().sort_values(ascending=False,by='tran_amount')) 
```

```python
#Columns =['age','gender','customer_city','customer_country','merchant_country','card_type','mcc_code','mar_status']
Agg = {'tran_amount': ['sum'], 'age': ['mean'],'gender':['mean'],'mar_status':['mean'],'mcc_code':['count'],'card_type':['max']}
res = df_complete.groupby('customer_id').agg(Agg)
res.head(100)
```

```python
print(res['age'])
```

```python
from sklearn.cluster import KMeans

#Age and spending Score'''
X1 = res[['age' , 'tran_amount']].iloc[: , 0:2].values
inertia = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
print(inertia)

plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

```python
algorithm = (KMeans(n_clusters = 4 ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
algorithm.fit(X1)
labels1 = algorithm.labels_
centroids1 = algorithm.cluster_centers_
```
