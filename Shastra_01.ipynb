```python
df_customer_1 = df_customer_1.drop(['Mark'],axis=1)
df_customer_1 = df_customer_1.rename(columns={'City_group': 'city_group'})

df_customer_1.head()
```

```python
df_basedata_0 = pd.merge(df_transaction_3, df_customer_1, left_on ='customer_id',right_on ='customer_id', how='left')
```

```python
#drop card_id_x, card_id_y, city_group_y

df_basedata_1 = df_basedata_0.drop(['card_id_x', 'card_id_y', 'city_group_y'],axis=1)

#rename city_group_x 
df_basedata_1 = df_basedata_1.rename(columns={'city_group_x': 'city_group'})
df_basedata_1.head()
```

#### Aggregations

```python
#aggregate 1
Agg = {'age': ['mean'],'dummy_gender': ['max'],'dummy_marital': ['mean'],'dummy_cardtype': ['max'],'tran_amount': ['sum'],'tran_id': ['count'], 'mcc_code':['nunique'], 'merchant_city':['nunique'], 'dummy_weekend':['sum'] }
df_basedata_agg1 = df_basedata_1.groupby('customer_id').agg(Agg).reset_index()
df_basedata_agg1.head(10)
```

```python
for col in df_basedata_agg1.columns: 
    print(col)
```

```python
df_basedata_agg1.info()
```

```python
# Group the data by month, and take the sum for each group (i.e. each month)
df_basedata_agg2 = df_basedata_1.set_index('tran_date').groupby('customer_id')["tran_amount"].resample("M").sum()
#df.groupby(['name', pd.Grouper(key='date', freq='M')])['ext price'].sum()
#df['month_year'] = pd.to_datetime(df['birth_date']).dt.to_period('M')
df_basedata_agg2.head()
```

```python
type(df_basedata_agg2)
```

```python
df_basedata_1['month_year'] = pd.to_datetime(df_basedata_1['tran_date']).dt.to_period('M')
df_basedata_1.head()
```

```python
#aggregate 2
Agg2 = {'tran_amount': ['sum'] }
df_basedata_agg2 = df_basedata_1.groupby(['customer_id','month_year']).agg(Agg2).reset_index()
df_basedata_agg2.head(10)
```

```python
type(df_basedata_agg2)
```

```python
df_basedata_agg2.info()
```

```python
#crosstab per customer by months
df_basedata_agg2_ct = pd.crosstab(df_basedata_1.customer_id, df_basedata_1.month_year ,values=df_basedata_1.tran_amount, aggfunc='sum').round(0).reset_index()
df_basedata_agg2_ct.head()
```

```python
df_basedata_agg2_ct.fillna(0,inplace = True)
df_basedata_agg2_ct.head()
```

```python
#aggregate 3

df_basedata_agg3_ct = pd.crosstab(df_basedata_1.customer_id, df_basedata_1.product_category ,values=df_basedata_1.tran_amount, aggfunc='sum').round(0).reset_index()
df_basedata_agg3_ct.head()
```

#### Merger Aggregation

```python
df_basedata_agg_total = pd.merge(df_basedata_agg1, df_basedata_agg2_ct, left_on ='customer_id',right_on ='customer_id', how='left')
df_basedata_agg_total.head()
```

```python
df_basedata_agg_total1 = pd.merge(df_basedata_agg_total, df_basedata_agg3_ct, left_on ='customer_id',right_on ='customer_id', how='left')
df_basedata_agg_total1.head()
```
